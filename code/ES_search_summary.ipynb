```python
import csv
with open('summary2.csv', 'rb') as csvfile:
    reader = csv.DictReader(csvfile)
    for inx, item in enumerate(reader):
        #print(inx+1, item['summary'])
        title = "TF-IDF summary{0}.csv".format(inx+1)
        print(title)
        label = "rank, pmcid, score"
        print(label)
        target = open(title, 'w')
        target.write(label + '\n')
        query = item['summary']
        res = es.search(index="tfidf",doc_type="article", q=query, analyzer="my_TFIDF_analyzer",size=100)
        count = 0
        for doc in res['hits']['hits']:
            count += 1
            text = '{0}, {1}, {2}'.format(count,doc['_source']['pmcid'],doc['_score'] )
            print (text)
            target.write(text + '\n')
        target.close()
```
